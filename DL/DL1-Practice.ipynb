```python
import pandas as pd
```

```python
df = pd.read_csv('HousingData.csv')
df.head(10)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.corr()['MEDV'].sort_values() #pearson correlation matrix
```

```python
X = df.loc[:, ['LSTAT', 'PTRATIO', 'RM']]
Y = df.loc[:, ['MEDV']]
X.shape, Y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=10)
X_train = X_train.fillna(X_train.mean())
X_test = X_test.fillna(X_test.mean())
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
#Input Layer
model.add(Dense(128, input_shape=(3,), activation='relu', name='input'))

#Middle Layer
model.add(Dense(64, activation='relu', name='layer-1'))

#Output Layer
model.add(Dense(1, activation='linear', name='output'))

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=100, validation_split=0.05)
```

```python
output = model.evaluate(X_test, y_test)
```

```python
print(f"Mean Squared Error: {output[0]}"
      ,f"Mean Absolute Error: {output[1]}",sep="\n")
```

```python
y_pred = model.predict(X_test)
```
